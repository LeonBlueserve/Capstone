Amsterdam has known many neighbourhood layouts in the years since then. After our Golden Age (17th century), Amsterdam started with population screening and dividing the city in several area’s (called Buurten or civilian neighbourhoods). Statistical data was collected including birth and mortality rates, average income, size of houses, deceases, population, etc. This data was the basis for what was known as personal taxation and income stratification. Each area or neighbourhood could have its own set of taxation rules. As a consequence rich and poor areas were created simply because people were moving to certain areas to avoid taxation (the poor).  The clustering created new dynamics in housing, rents, religion or social structure as a whole. To the present day neighbourhoods are still uniquely identifiable by its origin, if not due the name of the neighbourhood such as the Dutch-Indian Neighbourhood, the Golden-corner, the Sarphati (Portugese-area) or the Jewish-area then due to the fact that people from similar migration background seem to cluster together in certain neighbourhoods.
